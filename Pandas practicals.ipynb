# Group By in pandas

```python
data=pd.read_excel("ESD.xlsx")
```

```python
data.head(6)
```

```python
#No. of people accross different departments
gp=data.groupby("Department").agg({"EEID":"count"})
```

```python
gp
```

```python
#No. of males and females accross different departments
gp=data.groupby(["Department","Gender"]).agg({"EEID":"count"})
```

```python
gp
```

```python
#average annual salary of people from different countries
gp1=data.groupby("Country").agg({"Annual Salary":"mean"})
```

```python
gp1
```

```python
#maximum salary of workers by country
gp1=data.groupby("Country").agg({"Annual Salary":"max"})
```

```python
gp1
```

```python
#maximum salary of males and females in different countries
gp1=data.groupby(["Country","Gender"]).agg({"Annual Salary":"max"})
```

```python
gp1
```

```python
#maximum annual salary  and minimum age of males and females from different countries
gp1=data.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"min"})
```

```python
gp1
```

# Merge,concatenate and join

```python
data1={"Emp id":["E01","E02","E03","E04","E05","E06"],
      "Name":["Ram","Shyam","Rahul","Vishal","Ravi","John"],
      "Age":[35,57,46,23,26,36]}

data2={"Emp id":["E01","E02","E03","E04","E08","E09"],
       "Salary":[45000,56000,34000,30000,50000,62000]}
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
#it contains the values where 'Emp id' is similar
pd.merge(df1,df2,on="Emp id")
```

```python
#it contains all values of emp id from left dataframe(df1)
pd.merge(df1,df2,on="Emp id",how="left")
```

```python
#it contains all values of emp id from right dataframe(df2)
pd.merge(df1,df2,on="Emp id",how="right")
```

```python
data1={"Emp id":["E01","E02","E03","E04","E05","E06"],
      "Name":["Ram","Shyam","Rahul","Ankit","Ravi","John"],
      "Age":[35,57,46,23,26,36]}

data2={"Emp id":["E07","E08","E09","E010","E011","E012"],
       "Name":["Peter","Chris","Shivam","Raj","Aakash","Pawan"],
       "Age":[45,56,34,30,50,62]}
```

# Pivoting and Melting DataFrames

#### PIVOTING

```python
dict1={"keys":["k1","k2","k1","k2"],
      "Names":["John","Ben","David","Peter"],
      "Houses":["red","blue","green","red"]}
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

```python
df.pivot("keys","Names","Houses")   #index,columns,values
```

```python
dict1={"keys":["k1","k2","k1","k2"],
      "Names":["John","Ben","David","Peter"],
      "Houses":["red","blue","green","red"],
      "Grades":["3rd","8th","9th","8th"]}
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

```python
df.pivot(index="keys",columns="Names",values=["Houses","Grades"])
```

#### MELTING

```python
dict1={"Names":["John","Ben","David","Peter"],
      "Houses":["red","blue","green","red"],
      "Grades":["3rd","8th","9th","8th"]}
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

```python
pd.melt(df,id_vars=["Names"],value_vars=["Houses"])
```

```python
pd.melt(df,id_vars=["Names"],value_vars=["Houses","Grades"])
```

```python
pd.melt(df,id_vars=["Names"],value_vars=["Houses","Grades"],var_name="House&Grades",value_name="values")
```
